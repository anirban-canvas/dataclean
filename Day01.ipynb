```python
import pandas as pd
import numpy as np

df=pd.read_csv("data.csv")
df.head()
```

```python
dataC=['Corporate Author','Corporate Contributors',
      'Corporate Author',
    'Former owner','Engraver']
df.drop(dataC,inplace=True,axis=1)
df.head()
```

```python
df.Identifier.is_unique
```

```python
df.set_index('Identifier',inplace=True)
```

```python
df.head()
```

```python
extract=df['Date of Publication'].str.extract(r'^(\d{4})',expand=False)
df['Date of Publication']=pd.to_numeric(extract)
df.head()
```

```python
df['Place of Publication'].head(10)
```

```python
pubs=df['Place of Publication']
london=pubs.str.contains('London')
oxford=pubs.str.contains('Oxford')
df['Place of Publication'] = np.where(london,'Landon',
                                      np.where(oxford,'Oxford',
                                         pubs.str.replace('-',' ')))
df.head(10)
```
